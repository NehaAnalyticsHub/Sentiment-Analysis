```python
import torch
import torch.nn as nn
import numpy as np

class secondpart(nn.Module):
    def __init__(self, text_features_dim, hidden_dim, num_classes):
        #calls the initialization function of the parent class, nn.Module
        super(secondpart, self).__init__()
        #sets the dimensionality of the text features
        self.text_features_dim = text_features_dim
        #sets the number of hidden units
        self.hidden_dim = hidden_dim
        
        # Define layers
        self.text_hidden_layer = nn.Linear(text_features_dim, hidden_dim)
        #defines the first hidden layer that takes in the text hidden activations and produces new hidden activations
        self.hidden_layer1 = nn.Linear(hidden_dim, hidden_dim)
        #defines the second hidden layer
        self.hidden_layer2 = nn.Linear(hidden_dim, hidden_dim)
        #efines the third hidden layer.
        self.hidden_layer3 = nn.Linear(hidden_dim, hidden_dim)
        #defines the fourth hidden layer
        self.hidden_layer4 = nn.Linear(hidden_dim, hidden_dim)
        #defines the output layer which takes in the final hidden activations and produces a probability distribution over a set of classes
        self.output_layer = nn.Linear(hidden_dim, num_classes) # num_classes is the number of output classes
        
    def forward(self, text_features):
        #applies the text hidden layer to the input.
        text_hidden = self.text_hidden_layer(text_features)
        #applies the ReLU activation function to the text hidden activations.
        text_hidden = nn.functional.relu(text_hidden)
            
        # Pass through additional hidden layers
        #applies the first hidden layer to the text hidden activations
        hidden1 = self.hidden_layer1(text_hidden)
        #applies the ReLU activation function to the first hidden activations
        hidden1 = nn.functional.relu(hidden1)
        
        #here in other hidden layers , har layer sei piche wali us pe apply hho rhei hai or reu activation function lg rha 
        hidden2 = self.hidden_layer2(hidden1)
        hidden2 = nn.functional.relu(hidden2)
        
        hidden3 = self.hidden_layer3(hidden2)
        hidden3 = nn.functional.relu(hidden3)
        
        hidden4 = self.hidden_layer4(hidden3)
        hidden4 = nn.functional.relu(hidden4)
        
        # Pass through output layer
        #ye output layer pe pochnch gae hain
        output = self.output_layer(hidden4)
        
        return output
    

# Preprocess input data

#initializes an empty dictionary word_to_idx to map words to indices
word_to_idx = {}
#each caption or phir ech word pe loop
for caption in captions:
    for word in caption:
        #agr word dictionary mei nhi hai tou add krna 
        if word not in word_to_idx:
            #maps the word to a new index in dictionary
            word_to_idx[word] = len(word_to_idx)
            
#number of unique words from which we make dictionary
vocab_size = len(word_to_idx)
#creating tension mei dimensions and make it all zero
captions_tensor = torch.zeros(len(captions), len(word_to_idx))

#This loop iterates over each caption in captions and for each word in the caption, sets the corresponding element in 
#captions_tensor to 1.0. This converts each caption into a one-hot encoded tensor representation.
for i, caption in enumerate(captions):
    for word in caption:
        captions_tensor[i, word_to_idx[word]] = 1.0
        
#labels tensor from list conating word sentiment whoch we read from dataframe           
labels_tensor = torch.tensor(word_sentiment_list)



# Normalize captions tensor
#This converts the captions_tensor tensor to a numpy array and then normalizes each row of the array to have unit length using 
#numpy's linalg.norm() function. This creates a matrix of normalized caption embeddings.
captions_array = captions_tensor.numpy()
captions_array /= np.linalg.norm(captions_array, axis=-1, keepdims=True)
captions_array = captions_tensor.numpy()
captions_array /= np.linalg.norm(captions_array, axis=-1, keepdims=True)

# Convert labels to tensor
labels = torch.Tensor(word_sentiment_list)

# Define loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

text_features_dim = captions_array.shape[-1] 

# Train model
num_epochs = 10
for epoch in range(num_epochs):
    running_loss = 0.0
    for i in range(len(captions)):
        # Clear gradients
        optimizer.zero_grad()

        # Forward pass
        output = model(captions_tensor[i], image_features_tensor[i], text_features_dim, image_features_dim=image_features_tensor.shape[1])
        loss = criterion(output, labels[i])

        # Backward pass
        #which calculates the gradients of the loss with respect to the model parameters. These gradients are then used to 
        #update the model parameters using optimizer.step()
        loss.backward()
        #method that is used to update the parameters of the model based on the gradients computed during backpropagation.
        #It updates the parameters with the gradients accumulated in the optimizer (which are computed during loss.backward()
        #After the gradients have been computed, optimizer.step() adjusts the weights of the model in the direction that 
        #minimizes the loss function, thereby improving the model's performance
        optimizer.step()
 
        #The loss for the current batch is added
        running_loss += loss.item()

    # Print average loss for the epoch
    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, running_loss/len(captions)))
 
```
